```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from riskparity import design_pf
```

```python
assets = ['SPY', 'VWO','TLT', 'VNQ', "GLD", "KIM"]
```

```python
data = yf.download(assets)
data_cl = data['Adj Close'].dropna(axis=0)
```

```python
plt.plot(data_cl)
```

```python
k = len(assets)

def naive_rp(data_slice):
    k = data_slice.shape[1]
    asset_risk_budget = [1 / k] * k
    cov = data_slice.cov()
    x0 = np.ones(k) / k
    w = design_pf(cov.values, asset_risk_budget, x0)
    return w
```
